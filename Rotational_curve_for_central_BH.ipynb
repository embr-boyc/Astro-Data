```python
import astropy.units as u
import astropy.constants as ac
import numpy as np
```

How to define a function?

```python
def calculatingOrbitalVelocity(M, R):
    """
    This is a fuction that calculates the orbital velocity
    Input:
    M, the mass of the central object
    R, the orbital radius of the satellite
    Output:
    orbitalVelocity, the orbital velocity of the satellite
    """
    orbitalVelocity = np.sqrt(ac.G * M / R)
    return orbitalVelocity
```

```python
# Below we define input M and R
M = 1 * u.earthMass
R = 1 * u.earthRad

# Calculating the orbital velocity of an earth sattelite
v_earthSatellite = calculatingOrbitalVelocity(M, R)
print(v_earthSatellite.to(u.km/u.s))
```

We are going to introduce the concept of an array

```python
# orbital radius of the eight solar system planest from inside out in AU
# Mercury, Venus, Earth, Mars, Jupiter, Saturn, Uranus, Neptune
R_planet_arr = np.array([0.4, 0.7, 1.0, 1.5, 5.2, 9.6, 19.2, 30.1]) * u.au
```

```python
# Calculate the orbital velocities for the eight planets
M_sun = 1 * u.solMass
v_planet_arr = calculatingOrbitalVelocity(M_sun, R_planet_arr)
print(v_planet_arr.to(u.km/u.s))
```

```python
# prompt: can you help me with plotting with v_planet_arr and R_planet_arr

import matplotlib.pyplot as plt # Importing plotting package

plt.plot(R_planet_arr, v_planet_arr) # Plotting valocity vs orbital radius
plt.xlabel("Orbital Radius (AU)".format(R_planet_arr.unit)) # Plotting x axis label
plt.ylabel("Orbital Velocity (km/s)".format(v_planet_arr.unit)) # Plotting y axis label
plt.title("Orbital Velocity vs. Radius for Solar System Planets")
plt.show()
```

```python
R_suns_arr = np.arange(1, 30) * u.kpc
print(R_suns_arr.to(u.km))
```

```python
# Calculate the orbital velocities for the suns
M_blackhole = 8.546733E36 * u.kg
v_suns_arr = calculatingOrbitalVelocity(M_blackhole, R_suns_arr)
print(v_suns_arr.to(u.km/u.s))
```
